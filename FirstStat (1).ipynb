```python

import pandas as pd

profile_data = pd.read_csv("/mnt/mainstore/data/QiWi/Social_Demography/qiwi_users_profile_data_train.csv", encoding = 'cp1251', sep = '\t', delimiter = ';')

payment_data = pd.read_csv("/mnt/mainstore/data/QiWi/Social_Demography/qiwi_money_out_data_train.csv", encoding = 'cp1251', sep = '\t', delimiter = ';')
profile_data
```

```python
#wo_ids = profile_data[profile_data.sex == 'Жен']
#men_ids = profile_data[profile_data.sex == 'Муж']
```

```python
united = pd.merge(profile_data, payment_data, on="user_id", how="right")
united
```

```python
import matplotlib.pyplot as plt
import numpy as np

women = united[united.sex == "Жен"]
plt.hist(women[women.payment<10000].payment, 30, normed=True)
print("average ", np.average(women.payment))
print("median ", np.median(women.payment))
print("per person ", (len(women.payment))/(len(np.unique(women.user_id))))
```

```python
men = united[united.sex == "Муж"]
plt.hist(men[men.payment<10000].payment, 30, normed=True)
print("average ", np.average(men.payment))
print("median ", np.median(men.payment))
print("per person ", (len(men.payment))/(len(np.unique(men.user_id))))
```

```python
categorys = np.unique(payment_data.category)

for i in range(len(categorys)):
    curcat = categorys[i]
    print("\t\t\tCategory: ", curcat)
    print("Man average: ", np.average(men[men.category == curcat].payment))
    print("Man median: ", np.median(men[men.category == curcat].payment))
    print("Averege ammount of purches ", (len(men[men.category == curcat].payment))/
          (len(np.unique(men[men.category == curcat].user_id))))
    print("Popularity ", (len(men[men.category == curcat].payment))/(len(men.payment)))
    print("Amount of purchase ", len(men[men.category == curcat].payment))
    print("\n")
    print("Woman average: ", np.average(women[women.category == curcat].payment))
    print("Woman median: ", np.median(women[women.category == curcat].payment))
    print("Averege ammount of purchase", (len(women[women.category == curcat].payment))/
          (len(np.unique(women[women.category == curcat].user_id))))
    print("Popularity", (len(women[women.category == curcat].payment))/(len(women.payment)))
    print("Amount of purchase ", len(women[women.category == curcat].payment))
    print("\n\n")
```

```python
men_difference = []
for i in np.unique(men.user_id):
    men_difference.append(np.max(men[men.user_id == i].payment) - np.min(men[men.user_id == i].payment))
plt.hist([men_difference[i] for i in range(len(men_difference)) if men_difference[i]<50000], 100, normed = True)
plt.show()
```

```python
women_difference = []
for i in np.unique(women.user_id):
    women_difference.append(np.max(women[women.user_id == i].payment) - np.min(women[women.user_id == i].payment))
plt.hist([women_difference[i] for i in range(len(women_difference)) if women_difference[i]<50000], 100, normed = True)
plt.show()
```
